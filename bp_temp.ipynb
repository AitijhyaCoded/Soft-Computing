```python
# Fuzzy Logic using Triangular Membership Functions

def trimf(x, a, b, c):
    """Triangular membership: a=left foot, b=peak, c=right foot"""
    if x <= a or x >= c:
        return 0.0
    elif a < x <= b:
        return (x - a) / (b - a)
    else:
        return (c - x) / (c - b)

# BP membership (in mmHg)
def bp_membership(bp):
    low    = trimf(bp, 40, 60, 80)    # peaks at 60
    normal = trimf(bp, 60, 80, 100)   # peaks at 80
    high   = trimf(bp, 80, 100, 120)  # peaks at 100
    return low, normal, high

# Temperature membership (in Fahrenheit)
def temp_membership(temp):
    low    = trimf(temp, 93, 95, 98)      # peaks at 95
    normal = trimf(temp, 95, 98.6, 101)   # peaks at 98.6
    high   = trimf(temp, 99, 101, 104)    # peaks at 101
    return low, normal, high

# User input
bp   = float(input("Enter Blood Pressure (e.g. 75, 90, 110): "))
temp = float(input("Enter Temperature in F (e.g. 96, 98.6, 102): "))

# Fuzzify
bp_low, bp_normal, bp_high     = bp_membership(bp)
temp_low, temp_normal, temp_high = temp_membership(temp)

print(f"\nBP   -> Low: {bp_low:.2f}, Normal: {bp_normal:.2f}, High: {bp_high:.2f}")
print(f"Temp -> Low: {temp_low:.2f}, Normal: {temp_normal:.2f}, High: {temp_high:.2f}")

# Apply rules (AND = min)
rule1 = min(bp_high, temp_high)      # health = poor
rule2 = min(bp_normal, temp_normal)  # health = good
rule3 = min(bp_low, temp_normal)     # health = normal

print(f"\nRule 1 - Poor  : {rule1:.2f}")
print(f"Rule 2 - Good  : {rule2:.2f}")
print(f"Rule 3 - Normal: {rule3:.2f}")

# Defuzzify (weighted average)
poor_val, normal_val, good_val = 25, 50, 75

numerator   = (rule1 * poor_val) + (rule2 * good_val) + (rule3 * normal_val)
denominator = rule1 + rule2 + rule3

if denominator == 0:
    print("\nNo rules fired. Try different input values.")
else:
    score = numerator / denominator
    print(f"\nHealth Score: {score:.2f}")
    if score < 40:
        print("Health Status: POOR")
    elif score < 60:
        print("Health Status: NORMAL")
    else:
        print("Health Status: GOOD")
```
