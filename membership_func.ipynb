```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def triangular(x, a, b, c):
    if a <= x <= b:
        return (x-a)/(b-a)
    elif b < x <= c:
        return (c-x)/(c-b)
    return 0.0
    # return np.maximum(0.0, np.minimum((x-a)/(b-a),(c-x)/(c-b)))
x = np.linspace(0, 10, 1000)
a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))
v_triangular = np.vectorize(triangular)
y = v_triangular(x, a, b, c)
# y = triangular(x, a, b, c)
plt.plot(x, y, label=f"Triangular MF ({a}, {b}, {c})")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Triangular Membership Function")
plt.grid()
plt.legend()
plt.show()
```

```python
def trapezoidal(x, a, b, c, d):
    if a <= x <= b:
        return (x-a)/(b-a)
    elif b < x <= c:
        return 1.0
    elif c < x <= d:
        return (d-x)/(d-c)
    return 0.0
x = np.linspace(0, 10, 1000)
a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))
d = float(input("Enter d: "))
v_trapezoidal = np.vectorize(trapezoidal)
y = v_trapezoidal(x, a, b, c, d)
plt.plot(x, y, label=f"Trapezoidal MF ({a}, {b}, {c}, {d})")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Trapezoidal Membership Function")
plt.grid()
plt.legend()
plt.show()
```

```python
def generalised_bell(x, a, b, c):
    return 1 / (1 + ((x-c)/a)**(2*b))
x = np.linspace(0, 10, 1000)
a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))
v_generalised_bell = np.vectorize(generalised_bell)
y = v_generalised_bell(x, a, b, c)
plt.plot(x, y, label=f"Generalised Bell MF ({a}, {b}, {c})")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Generalised Bell Membership Function")
plt.grid()
plt.legend()
plt.show()
```

```python
def gaussian(x, sigma, c):
    return np.exp(-0.5 * ((x-c)/sigma)**2)
x = np.linspace(0, 10, 1000)
c = float(input("Enter centre: "))
sigma = float(input("Enter width: "))
v_gaussian = np.vectorize(gaussian)
y = v_gaussian(x, sigma, c)
plt.plot(x, y, label=f"Gaussian MF ({sigma}, {c})")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gaussian Membership Function")
plt.grid()
plt.legend()
plt.show()
```

```python
def sigmoidal(x, a, c):
    return 1/(1+np.exp(-a*(x-c)))
x = np.linspace(0, 10, 1000)
a = float(input("Enter a: "))
c = float(input("Enter c: "))
v_sigmoidal = np.vectorize(sigmoidal)
y = v_sigmoidal(x, a, c)
plt.plot(x, y, label=f"Sigmoidal MF ({a}, {c})")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Sigmoidal Membership Function")
plt.grid()
plt.legend()
plt.show()
```
