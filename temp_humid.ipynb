```python
# Fuzzy Union: Acceptable Temp OR Acceptable Humidity

# T = membership/temperature
T = {16: 0.4, 18: 0.8, 20: 1.0, 22: 1.0, 24: 0.8, 26: 0.5}

# H = membership/humidity
H = {0: 0.2, 20: 0.8, 40: 1.0, 60: 0.6, 80: 0.2}

# Union = max of memberships for common elements, keep rest as is
all_keys = set(T.keys()) | set(H.keys())  # all unique elements from both

union = {}
for k in sorted(all_keys):
    t_val = T.get(k, 0)  # if key not in T, treat as 0
    h_val = H.get(k, 0)  # if key not in H, treat as 0
    union[k] = max(t_val, h_val)

print("Fuzzy Union (T ∪ H) - Acceptable Temp OR Acceptable Humidity:")
for k, v in union.items():
    print(f"  {v}/{k}")
```
