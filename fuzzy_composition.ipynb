```python
def max_min_composition(R, S):
    res_max = []
    for i in range(len(R)):
        res_max.append([])
        for j in range(len(S[0])):
            res_min = []
            for k in range(len(R[0])):
                res_min.append(min(R[i][k], S[k][j]))
            res_max[i].append(max(res_min))
    return res_max
R = [[0.5, 0.6, 0.3], [0.4, 0.8, 0.7]]
S = [[0.5, 0.6], [0.4, 0.8], [0.3, 0.8]]
for i in max_min_composition(R, S):
    print(i)
```

```python
def max_pdt_composition(R, S):
    res_max = []
    for i in range(len(R)):
        res_max.append([])
        for j in range(len(S[0])):
            res_pdt = []
            for k in range(len(R[0])):
                res_pdt.append(R[i][k]*S[k][j])
            res_max[i].append(round(max(res_pdt), 2))
    return res_max
R = [[0.5, 0.6, 0.3], [0.4, 0.8, 0.7]]
S = [[0.5, 0.6], [0.4, 0.8], [0.3, 0.8]]
for i in max_pdt_composition(R, S):
    print(i)
```
