```python
# Defuzzification Methods
# User inputs fuzzy set as (membership, domain_value) pairs

n = int(input("How many elements in your fuzzy set? "))

membership = []
domain     = []

for i in range(n):
    x  = float(input(f"Enter domain value {i+1}  : "))
    mu = float(input(f"Enter membership value {i+1}: "))
    domain.append(x)
    membership.append(mu)

print(f"\nYour fuzzy set: { {domain[i]: membership[i] for i in range(n)} }")

# ── 1. CENTROID OF AREA (COA) ─────────────────────────
# Formula: sum(x * mu) / sum(mu)
def centroid_of_area(domain, membership):
    numerator   = sum(domain[i] * membership[i] for i in range(len(domain)))
    denominator = sum(membership)
    return numerator / denominator

# ── 2. BISECTOR OF AREA (BOA) ─────────────────────────
# Find x where left area = right area
def bisector_of_area(domain, membership):
    total_area = sum(membership)          # total area
    half_area  = total_area / 2

    cumulative = 0
    for i in range(len(domain)):
        cumulative += membership[i]
        if cumulative >= half_area:       # found the bisector point
            return domain[i]

# ── 3. MEAN OF MAXIMUM (MOM) ──────────────────────────
# Average of all domain values that have maximum membership
def mean_of_maximum(domain, membership):
    max_mu    = max(membership)
    max_points = [domain[i] for i in range(len(domain)) if membership[i] == max_mu]
    return sum(max_points) / len(max_points)

# ── 4. SMALLEST OF MAXIMUM (SOM) ─────────────────────
# Smallest domain value with maximum membership
def smallest_of_maximum(domain, membership):
    max_mu = max(membership)
    for i in range(len(domain)):
        if membership[i] == max_mu:
            return domain[i]             # return first (smallest) occurrence

# ── RESULTS ───────────────────────────────────────────
print(f"\nCentroid of Area  (COA): {centroid_of_area(domain, membership):.2f}")
print(f"Bisector of Area  (BOA): {bisector_of_area(domain, membership):.2f}")
print(f"Mean of Maximum   (MOM): {mean_of_maximum(domain, membership):.2f}")
print(f"Smallest of Maximum(SOM): {smallest_of_maximum(domain, membership):.2f}")
```
