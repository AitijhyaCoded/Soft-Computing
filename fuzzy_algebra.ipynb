```python
A = {"x1": 0.5, "x2": 0.7, "x3": 0.3}
B = {"x1": 0.8, "x2": 0.2, "x3": 0.5, "x4": 0.6}
```

```python
def algebraic_sum(A, B):
    univ = set(A.keys()) | set(B.keys())
    res = {}
    for element in univ:
        mu_A = A.get(element, 0)
        mu_B = B.get(element, 0)
        mu_A_plus_B = mu_A + mu_B - mu_A * mu_B
        res[element] = mu_A_plus_B
    return res
print(algebraic_sum(A, B))
```

```python
def algebraic_diff(A, B):
    univ = set(A.keys()) | set(B.keys())
    res = {}
    for element in univ:
        mu_A = A.get(element, 0)
        mu_B = B.get(element, 0)
        mu_A_minus_B = mu_A * (1 - mu_B)
        res[element] = mu_A_minus_B
    return res
print(algebraic_diff(A, B))
```

```python
def cartesian_pdt(A, B):
    res = {} 
    for x in A.keys():
        for y in B.keys():
            if x!=y:
                res[(x, y)] = min(A[x], B[y])
    return res
A = {"x1": 0.5, "x2": 0.7}
B = {"y1": 0.8, "y2": 0.2, "y3": 0.5}
result = cartesian_pdt(A, B)
for i in result:
    print(i, ":", result[i])
```
