```python
import math

# Generalized Bell membership function
# Formula: mu = 1 / (1 + |(x - c) / a| ^ (2b))
def bell(x, a, b, c):
    return 1 / (1 + abs((x - c) / a) ** (2 * b))

x = float(input("Enter value of x (age): "))

# Base memberships
mu_young = bell(x, 20, 2, 0)    # bell(x, a=20, b=2, c=0)
mu_old   = bell(x, 30, 3, 100)  # bell(x, a=30, b=3, c=100)

print(f"\nmu_young = {mu_young:.4f}")
print(f"mu_old   = {mu_old:.4f}")

# ── 1. MORE OR LESS YOUNG ─────────────────────────────
# "More or less" = sqrt = power of 0.5 (dilution)
more_or_less_young = mu_young ** 0.5
print(f"\n1. More or Less Young : {mu_young:.4f}^0.5 = {more_or_less_young:.4f}")

# ── 2. NOT YOUNG AND NOT OLD ──────────────────────────
# NOT = 1 - mu
# AND = min
not_young = 1 - mu_young
not_old   = 1 - mu_old
not_young_and_not_old = min(not_young, not_old)
print(f"\n2. Not Young and Not Old:")
print(f"   not_young = 1 - {mu_young:.4f} = {not_young:.4f}")
print(f"   not_old   = 1 - {mu_old:.4f}   = {not_old:.4f}")
print(f"   min({not_young:.4f}, {not_old:.4f}) = {not_young_and_not_old:.4f}")

# ── 3. YOUNG BUT NOT TOO YOUNG ────────────────────────
# "Too young" = concentration = power of 2
# "But" = AND = min
too_young          = mu_young ** 2
young_not_too_young = min(mu_young, 1 - too_young)
print(f"\n3. Young but Not Too Young:")
print(f"   too_young = {mu_young:.4f}^2 = {too_young:.4f}")
print(f"   min(young, not_too_young) = min({mu_young:.4f}, {1-too_young:.4f}) = {young_not_too_young:.4f}")

# ── 4. EXTREMELY OLD ──────────────────────────────────
# "Extremely" = concentration = power of 3
extremely_old = mu_old ** 3
print(f"\n4. Extremely Old : {mu_old:.4f}^3 = {extremely_old:.4f}")
```
