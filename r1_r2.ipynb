```python
# Fuzzy Max-Min and Max-Product Composition

x = [1,2,3]
y = ['a','b','c','d']
z = ["one","two"]

# R1: X -> Y (3 rows for x=1,2,3 and 4 cols for y=a,b,c,d)
R1 = [
# y = a    b    c    d
    [0.2, 0.8, 0.4, 0.6],  # x=1
    [0.5, 0.3, 0.9, 0.1],  # x=2
    [0.7, 0.6, 0.2, 0.8],  # x=3
]

# R2: Y -> Z (4 rows for y=a,b,c,d and 2 cols for z=one,two)
R2 = [
# z = one,two
    [0.4, 0.6],  # y=a
    [0.8, 0.2],  # y=b
    [0.3, 0.7],  # y=c
    [0.5, 0.9],  # y=d
]

x_index = x.index(3)   # x=3 is index '2' (0-based)
z_index = z.index("one")   # z=one is index '0'

# Extract row 3 from R1 and column "one" from R2
r1_row = R1[x_index]                        # [0.7, 0.6, 0.2, 0.8]
r2_col = [R2[y][z_index] for y in range(4)] # [0.4, 0.8, 0.3, 0.5]

print(f"R1 row (x=3)     : {r1_row}")
print(f"R2 col (z=one)   : {r2_col}")

# Max-Min Composition
min_values = [min(r1_row[y], r2_col[y]) for y in range(4)]
max_min_result = max(min_values)

print(f"\nMax-Min -> mins  : {min_values}")
print(f"Max-Min Result   : {max_min_result}")

# Max-Product Composition
product_values = [r1_row[y] * r2_col[y] for y in range(4)]
max_product_result = max(product_values)

print(f"\nMax-Product -> products : {product_values}")
print(f"Max-Product Result      : {max_product_result}")
```
